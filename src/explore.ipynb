Round times up or down, from 00 to 24.

```python
# Function to round time values to the nearest hour
def round_time(time_str):
    if isinstance(time_str, str):
        time_parts = time_str.split(':')
        hours = int(time_parts[0])
        minutes = int(time_parts[1].split('.')[0])  # Extract minutes before the decimal point
        if minutes >= 30:  # Round up to the next hour if minutes are 30 or more
            hours += 1
        return f"{hours:02d}:00"  # Ensure leading zero for single-digit hours
    else:
        return None  # Handle non-string values
# List of columns to round the time values
time_columns = ['1depTimeOnly', '1arrTimeOnly', '2depTimeOnly', '2arrTimeOnly', '3depTimeOnly', '3arrTimeOnly', '4depTimeOnly', '4arrTimeOnly']
# Apply the round_time function to each column
for column in time_columns:
    even_more_useful_feat_data[column] = even_more_useful_feat_data[column].apply(round_time)
# Display the DataFrame with rounded time values
print(even_more_useful_feat_data)
```

Add "waitingTime" column: difference between the two (or more) flights. For example if we have two segments: take the departure of the second flight and subtract the arrival of the first flight. This will give back the time the passenger has to wait between the two flights. 

- if numberOfSegments 2: 2depTime - 1arrTime
- if number 3: 2depTime - 1arrTime + 3depTime - 2arrTime
- etc

```python
even_more_useful_feat_data.head()
```

```python
# Function to calculate waiting time based on the number of segments
def calculate_waiting_time(row):
    if row['numberOfSegments'] == 2:
        return (parse_time(row['2depTimeOnly']) - parse_time(row['1arrTimeOnly'])).total_seconds() / 3600
    elif row['numberOfSegments'] == 3:
        return ((parse_time(row['2depTimeOnly']) - parse_time(row['1arrTimeOnly'])) +
                (parse_time(row['3depTimeOnly']) - parse_time(row['2arrTimeOnly']))).total_seconds() / 3600
    elif row['numberOfSegments'] == 4:
        return ((parse_time(row['2depTimeOnly']) - parse_time(row['1arrTimeOnly'])) +
                (parse_time(row['3depTimeOnly']) - parse_time(row['2arrTimeOnly'])) +
                (parse_time(row['4depTimeOnly']) - parse_time(row['3arrTimeOnly']))).total_seconds() / 3600
    # Add more conditions for higher number of segments if needed
    else:
        return None  # Handle cases with more than 4 segments
# Function to parse time strings into datetime objects
def parse_time(time_str):
    if time_str == '24:00':
        # Treat '24:00' as '00:00' of the next day
        return datetime.strptime('00:00', '%H:%M')
    else:
        return datetime.strptime(time_str, '%H:%M')
# Apply the function to create the "waitingTime" column
even_more_useful_feat_data['waitingTime'] = even_more_useful_feat_data.apply(calculate_waiting_time, axis=1)
# Display the DataFrame with the new "waitingTime" column
print(even_more_useful_feat_data)
```
